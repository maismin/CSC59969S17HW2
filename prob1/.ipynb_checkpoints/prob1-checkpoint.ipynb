```python
%matplotlib inline
import numpy as np
import matplotlib

import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
data = np.recfromcsv('crabs.csv')
data.dtype.fields
cl = data[:]['cl'] # x
cw = data[:]['cw'] # y
bd = data[:]['bd'] # z

print(round(cl.max()))
```

```python
# create figure
fig = plt.figure(figsize=(10,10))

# create grid of 1 row and 1 column
gs = mpl.gridspec.GridSpec(1,1)

# have only two spines visible and set properties
ax = fig.add_subplot(gs[0])
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

# set tick properties
ax.xaxis.set_tick_params('major', length=2, labelsize=12)
ax.yaxis.set_tick_params('major', length=2, labelsize=12)

# set axis limits
ax.set_xlim((0, round(cl.max()+20)))
ax.set_ylim((0, round(cw.max()+20)))

# scale view so show all datapoints
ax.autoscale_view()

# set labels
ax.set_xlabel('Carapace Length (mm)', fontsize=12, family='sans-serif')
ax.set_ylabel('Carapace Width (mm)', fontsize=12, family='sans-serif')

ax.set_title('Carapace Length vs Carapace Width')

# plot data
ax.plot(cl, cw, marker='.', markersize=4, linestyle='')
```

```python
# create figure
fig = plt.figure(figsize=(20,10))

# create grid of 1 row and 1 column
gs = mpl.gridspec.GridSpec(1,2)

# LEFT SUBPLOT ========================================
# have only two spines visible and set properties
ax = fig.add_subplot(gs[0])
ax1 = ax.twinx()
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

# set tick properties
ax.xaxis.set_tick_params('major', length=2, labelsize=12)
ax.yaxis.set_tick_params('major', length=2, labelsize=12)

# set axis limits
ax.set_xlim((0, round(cl.max()+20)))
ax.set_ylim((0, round(cw.max()+20)))

# scale view so show all datapoints
ax.autoscale_view()

# set labels
ax.set_xlabel('Carapace Length (mm)', fontsize=12, family='sans-serif')
ax.set_ylabel('Carapace Width (mm)', fontsize=12, family='sans-serif')
ax1.set_ylabel('Body Depth (mm)', fontsize=12, family='sans-serif')

# plot data
ax.plot(cl, cw, marker='.', markersize=4, color='r', linestyle='', label='Carapace Width')
ax1.plot(cl, bd, marker='x', markersize=4, color='b', linestyle='', label='Body Depth')
ax.legend(loc='upper left')
ax1.legend(loc='upper right')


# RIGHT SUBPLOT ========================================
# have only two spines visible and set properties
ax = fig.add_subplot(gs[1])
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

# set tick properties
ax.xaxis.set_tick_params('major', length=2, labelsize=12)
ax.yaxis.set_tick_params('major', length=2, labelsize=12)

# set axis limits
ax.set_xlim((0, round(cl.max()+10)))
ax.set_ylim((0, round(cw.max()+10)))

# scale view so show all datapoints
ax.autoscale_view()

# set labels
ax.set_xlabel('Carapace Length (mm)', fontsize=12, family='sans-serif')
ax.set_ylabel('Carapace Width (mm)', fontsize=12, family='sans-serif')

# plot data
ax.scatter(cl, cw, marker='o', s=(bd*15))
ax.legend(['Body Depth'])
```

I think that using two y-axis represents the data more accurately. The variable markersize on the right figure is harder to see.

```python
d = data[['sp','sex','cl','cw']]
dBlueMale = d[0:49]
dBlueFemale = d[50:99]

# create figure
fig = plt.figure(figsize=(10,10))

# create grid of 1 row and 1 column
gs = mpl.gridspec.GridSpec(1,1)

# have only two spines visible and set properties
ax = fig.add_subplot(gs[0])
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

# set tick properties
ax.xaxis.set_tick_params('major', length=2, labelsize=12)
ax.yaxis.set_tick_params('major', length=2, labelsize=12)

# set axis limits
ax.set_xlim((10, round(cl.max()+10)))
ax.set_ylim((10, round(cw.max())))

# scale view so show all datapoints
ax.autoscale_view()

# set labels
ax.set_xlabel('Carapace Length (mm)', fontsize=12, family='sans-serif')
ax.set_ylabel('Carapace Width (mm)', fontsize=12, family='sans-serif')

ax.set_title('Carapace Length vs Carapace Width')

# plot data
ax.plot(dBlueMale['cl'], dBlueMale['cw'], marker='.', markersize=10, linestyle='--', color='b')
ax.plot(dBlueFemale['cl'], dBlueFemale['cw'], marker='.', markersize=10, linestyle='--', color='r')
ax.legend(['Male','Female'])
```
